## <span style = "color:orange"> Visualization </span>

### Plot a scatterplot of Avg_credit_Limit vs Total_Credit_Cards with respect to Total_calls_made

```python
df.columns
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Avg_Credit_Limit',y='Total_Credit_Cards',data=df,hue='Total_calls_made',palette='Dark2')
plt.show()
```

### Plot correlation matix on Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

### Plot a pairplot

```python
sns.pairplot(df)
plt.show()
```

### Create K Means model (Let k = 4)

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3)
```

### Fit predict for our Dataset

```python
clusters=model.fit_predict(df)
```

```python
clusters
```

### Assign the predicted clusters to a new column in our dataset

```python
df['Clusters']=clusters
```

```python
df.head()
```

```python
df['Clusters'].unique()
```

## Visualize the clusters

```python
# Use x = 'Avg_Credit_Limit' ang y =  'Total_Credit_Cards'
```

```python
df0=df[df['Clusters']==0]
df1=df[df['Clusters']==1]
df2=df[df['Clusters']==2]
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Avg_Credit_Limit',y='Total_Credit_Cards',data=df0)
sns.scatterplot(x='Avg_Credit_Limit',y='Total_Credit_Cards',data=df1)
sns.scatterplot(x='Avg_Credit_Limit',y='Total_Credit_Cards',data=df2)
plt.show()
```
